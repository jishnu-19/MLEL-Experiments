```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

penguin = sns.load_dataset('penguins')
df = penguin.copy()
df.head()
```

```python
df.describe()
```

```python
species_counts = df['species'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar')
plt.title('Distribution of Penguin Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.plot(df['bill_length_mm'])
plt.title("Line Plot - Bill Length")
plt.xlabel("Sample Index")
plt.ylabel("Bill Length (mm)")
plt.show()
```

```python
species_counts = df['species'].value_counts().sort_index()

plt.figure(figsize=(8, 8))
species_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Distribution of Penguin Species')
plt.ylabel('')
plt.legend(labels=species_counts.index, loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
plt.hist(df['flipper_length_mm'])
plt.title("Histogram - Flipper Length")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['bill_length_mm'], df['bill_depth_mm'])
plt.title("Scatter Plot - Bill Dimensions")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.imshow(corr)
plt.colorbar()
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['bill_length_mm'])
plt.title('Box Plot of Bill Length (mm)')
plt.ylabel('Bill Length (mm)')
plt.show()
```

```python
plt.step(range(len(df)), df['flipper_length_mm'])
plt.title("Step Plot - Flipper Length")
plt.xlabel("Sample Index")
plt.ylabel("Flipper Length (mm)")
plt.show()
```
